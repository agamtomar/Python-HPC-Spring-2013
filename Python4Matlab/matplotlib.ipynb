```python
%pylab inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Creating some random data

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(100)
print x
```

## Simple line plot

```python
plt.plot(x,'-bs')
```

## Same with same LaTex text

```python
plt.plot(x, 'rs')
plt.xlabel('Smarts')
plt.ylabel('Probablity')
plt.title('Histogram of IQ')
plt.text(10,135, r'$\mu=100,\ \sigma=15$')
plt.grid(True)
```

## Other 2D plot Styles

Look at http://matplotlib.org/gallery.html for examples

```python
x = linspace(0, 5, 10)
xx = np.linspace(-0.75, 1., 100)
n = array([0,1,2,3,4,5])

fig, axes = subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*randn(len(xx)))

axes[1].step(n, n**2, lw=2)

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
```

## Generate a PDF with the pdf backend

Note: you need to restart the notebook as the backend can only be selected before pylab is imported

```python
import matplotlib
matplotlib.use('pdf')
import matplotlib.pylab as plt
import numpy
```

```python
t = numpy.linspace(0, 10, 100)
plt.plot(t, numpy.cos(t)*numpy.sin(t))
plt.savefig("test.pdf")
```
