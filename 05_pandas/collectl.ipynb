# Collectl time-series data

##Objectives

* Create an index with `datetime` objects
* Concatenate several dataframes together
* Pandas does the data alignment


```python
import os
import glob
import pandas as pd

files = glob.glob(os.path.join('collectl','data-node*'))

tmp = pd.read_table(files[0], header=True, skiprows=range(16), sep ='\s+')
tmp.index = pd.to_datetime(tmp.pop('Time'))
tmp.pop('#Date')
```

```python
data = {}
# Read in every file
for f in files:
    index = f.find('node')
    node_name = f[index:index+8]
    tmp = pd.read_table(f, header=True, skiprows=range(16), sep ='\s+')
    tmp.index = pd.to_datetime(tmp.pop('Time'))
    tmp.pop('#Date')
    
    # Create a dictionary item for each column
    for c in tmp.columns:
        dic_item = pd.DataFrame(tmp, columns=[c])
        dic_item.columns = [node_name]
        # Create the dictionary for the column if it doesn't exist
        if c not in data:
            data[c] = []
        data[c].append(dic_item)
```

```python
for i, v in enumerate(data):
    if v.find('CPU') > 0:
        print v
```
