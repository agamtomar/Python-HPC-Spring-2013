#DataFrame

* Dataframe is like a dictionary
* You can index column names or lists of column names
* avoid integer column names
* The first dimension is the index.  
* Subsequent dimensions are the columns

```python
df = pd.DataFrame({'One' : np.random.random(5), 'Two' : np.random.random(5)})
df
```

```python
df['One']
```

```python
df.ix[:3]
```

```python
df = pd.DataFrame({'a' : np.random.random(5), 'b':np.random.random(5), 'c' : np.random.random(5)}, 
     index=['one', 'two', 'three', 'four', 'five'])
df
```

```python
df.ix['two':'five']
```

```python
df[:3]
```

#Cross Section

```python
df.xs('three')
```

#Indexing by Datetime

```python
rng = pd.date_range('3/14/2013', periods=50, freq='T')
rng
```

```python
import random

d = {'pt_1':pd.Series(randn(len(rng)), index=rng), 'pt_2':pd.Series(randn(len(rng)), 
      index=rng)}
df = pd.DataFrame(d)
df.head()
```

```python
d = {'pt_1':pd.Series(randn(len(rng)), index=rng), 'pt_2':pd.Series(randn(len(rng)), index=rng),
     'counts':pd.Series(randint(10,size=len(rng)), index=rng)}
df = pd.DataFrame(d)
df.head()
```
