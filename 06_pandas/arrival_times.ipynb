##Questions
- Which month has the most delays?
- Which day of the week has the most delays?

```python
import os
import pandas as pd
import datetime as datetime
```

```python
den = pd.read_csv(os.path.join('arrival_times', 'den-2008.csv'))
```

```python
len(den)
```

```python
den.head()
```

```python
den_min = den[['Year', 'Month', 'DayofMonth','ArrTime','ArrDelay','DayOfWeek']]
```

```python
den_min = den_min.dropna()
```

```python
def create_datetime(df):
    ##hhmm 
    hour = int(df['ArrTime'])/100
    minutes = int(df['ArrTime']) - hour*100
    try:
        dt = datetime.datetime(year = int(df['Year']), 
                           month = int(df['Month']), 
                           day = int(df['DayofMonth']),
                           hour = hour,
                           minute = minutes)
    except:
        return datetime.datetime(year=2000, month=1, day=1)
    return dt
   
```

```python
tmp_index = den_min.apply(create_datetime, axis=1)
```

```python
den_min.index = tmp_index
```

Get rid of the dates we don't want

```python
bad_index = den_min.index > datetime.datetime(year=2000, month=1, day=1)
```

```python
bad_index
```

```python
den_min = den_min[bad_index]
```

Do we have any duplicates?

```python
den_min.index.is_unique
```

### Which month has the most delays?

```python
den_month = den_min.resample('M', how=mean)
```

```python
den_month['ArrDelay'].plot()
```

### Which day of the week has the most delays?

```python
per_day_year = den_min['ArrDelay'].resample('D', how=mean)
```

```python
per_day_year.plot()
```

```python
day_of_week = per_day_year.groupby(lambda x: x.weekday).mean()
```

```python
day_of_week
```

```python
day_of_week.plot()
```
