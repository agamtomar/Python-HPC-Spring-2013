## Assemble

```python
years = range(1880,2011)
name_ind = ['name','sex','births']
frames = []
```

```python
for year in years:
    filename = os.path.join('baby_names','yob'+str(year)+'.txt')
    tmp = pd.read_csv(filename, names=name_ind)
    tmp['year'] = year
    #frames.append(tmp)
    print tmp
```

```python
names = pd.concat(frames, ignore_index=True)
```

```python
names
```

```python
names.head()
```

```python
total_births = names.pivot_table('births', rows='year', cols='sex', aggfunc=sum)
```

```python
total_births.tail()
```

```python
total_births.plot()
```

## Apply

```python
grp = names.groupby(['sex'])
```

```python
for name, group in grp:
    print name
    print group
```

My `grp` has two DataFrames, `M` and `F`

```python
def top(df, n=10, column='births'):
    return df.sort_index(by=column).tail(n)
```

```python
top_names = top(names)
top_names.head()
```

```python
top_names.index
```

```python
top_names_by_gender = grp.apply(top)
top_names_by_gender
```

The `top` function is called on each `DataFrame` in `grp`.  Then the results are put back together using `pandas.concat`.

### Popular Names

```python
def add_prop(df):
    births = df.births.astype(float)
    total = births.sum()
    df['prop'] = births/total
    return df
```

```python
pop_names = names.groupby(['year','sex']).apply(add_prop)
```

```python
pop_names.head()
```

```python
import numpy as np
np.allclose(pop_names.groupby(['year','sex']).prop.sum(),1)
```

```python
def top1000names(df):
    return df.sort_index(by='births', ascending=False)[:1000]
```

```python
grp = pop_names.groupby(['year','sex'])
top_names = grp.apply(top1000names)
```

```python
top_names.head()
```

```python
top_names_prop = top_names.pivot_table('prop', rows='year', cols='sex', aggfunc=sum)
```

```python
top_names_prop.plot()
```
