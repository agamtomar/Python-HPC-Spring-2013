#Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
import style
```

##Histograms

```python
mu = 100
sigma = 15
x = mu + sigma * np.random.randn(300)
```

```python
fig, ax = plt.subplots()

ax.hist(x, 10, normed=1, facecolor='#377EB8', alpha=0.5)

style.ggplot(ax)
```

##Density

```python
from scipy import stats
```

```python
density = stats.kde.gaussian_kde(x)
```

```python
fig, ax = plt.subplots()

xd = np.linspace(min(x)-10, max(x)+10, 100)
ax.plot(xd, density(xd), color='#FF7F00')

style.ggplot(ax)
```

```python
fig, ax = plt.subplots()

ax.plot(xd, density(xd), color='#FF7F00')

ax.fill_between(xd, 0, density(xd), alpha=0.5, color='#FF7F00')

style.ggplot(ax)
```

##Combined

```python
fig, ax = plt.subplots()

ax.hist(x, 10, normed=1, facecolor='#377EB8', alpha=0.5)
ax.plot(xd, density(xd), lw=2, color='#FF7F00')

style.ggplot(ax)
```

##Compare distributions

```python
x1 = 100 + 15 * np.random.randn(100)
x2 = 50 + 10 * np.random.randn(100)

density1 = stats.kde.gaussian_kde(x1)
density2 = stats.kde.gaussian_kde(x2)

x = np.linspace(0, 150, 200)
```

```python
fig, ax = plt.subplots()

ax.fill_between(x, 0, density1(x), alpha=0.5, color='#1B9E77')
ax.fill_between(x, 0, density2(x), alpha=0.5, color='#D95F02')

ax.plot([100,100],[0,density1(100)], color ='#1B9E77', linewidth=1, linestyle="--")
ax.plot([50,50],[0,density2(50)], color ='#D95F02', linewidth=1, linestyle="--")

style.ggplot(ax)
```
