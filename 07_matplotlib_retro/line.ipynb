##Scatter with trendline

Get the cars data (again)

```python
cars = pd.read_csv(os.path.join('data','cars.csv'))

c4 = cars[cars['cyl'] == 4]
c6 = cars[cars['cyl'] == 6]
c8 = cars[cars['cyl'] == 8]
```

Get the trend line using `stats.linregress`.

```python
from scipy import stats

x = cars['wt'].values
y = cars['mpg'].values
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

ye = slope*x + intercept
```

Add to plot

```python
fig, ax = plt.subplots()
ax.plot(x, ye, label="estimate", c='black')

ax.scatter(c4['wt'], c4['mpg'], s=50, c='#FF7F00', edgecolors='none', label='4 cyl')
ax.scatter(c6['wt'], c6['mpg'], s=50, c='#4DAF4A', edgecolors='none', label='6 cyl')
ax.scatter(c8['wt'], c8['mpg'], s=50, c='#377EB8', edgecolors='none', label='8 cyl')



ax.set_xlabel('wt')
ax.set_ylabel('mpg')
ax.legend(loc='best', scatterpoints=1)

style.ggplot(ax)
```

## Text annotation with Latex

```python
lab = 'y = %.2f'%(slope) + '\\cdot x + ' + '%.2f'%(intercept)
```

```python
lab
```

```python
fig, ax = plt.subplots(figsize=(9.0, 6.0))

ax.plot(x, ye, label="estimate", c='black')

ax.scatter(c4['wt'], c4['mpg'], s=50, c='#FF7F00', edgecolors='none', label='4 cyl')
ax.scatter(c6['wt'], c6['mpg'], s=50, c='#4DAF4A', edgecolors='none', label='6 cyl')
ax.scatter(c8['wt'], c8['mpg'], s=50, c='#377EB8', edgecolors='none', label='8 cyl')

ax.set_xlabel('wt')
ax.set_ylabel('mpg')

ax.annotate(r'$' + lab + '$',
         xy=(3,30), 
         fontsize=16)

ax.legend(loc='best', scatterpoints=1)

style.ggplot(ax)
```
